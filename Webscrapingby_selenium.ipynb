```python
import csv
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
```

```python
# Setup WebDriver
options = webdriver.ChromeOptions()
options.headless = False  # Set to True if you want to run in headless mode

# Initialize WebDriver
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)
driver.get("https://www.iplt20.com/stats/2022")

# Wait for elements to load
wait = WebDriverWait(driver, 30)
wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, ".np-battingtable_contaner")))


# Extract match data
matches = driver.find_elements(By.CSS_SELECTOR, ".np-battingtable_contaner")

match_details = []
for match in matches:
    match_info = {}

    # Extract the date
    try:
        match_info['POS'] = match.find_element(By.XPATH, ".//div[contains(@class, 'st-table__head')]").text.strip()
    except:
        match_info['POS'] = 'N/A'

    # Extract the match number
    try:
        match_info['Player'] = match.find_element(By.XPATH, ".//span[contains(@class, 'vn-matchOrder')]").text.strip()
    except:
        match_info['Player'] = 'N/A'

    # Extract teams
    try:
        match_info['Runs'] = match.find_element(By.XPATH, ".//div[contains(@class, 'live-score')]").text.strip()
    except:
        match_info['Runs'] = 'N/A'

    # Extract match result
    try:
        match_info['Match'] = match.find_element(By.XPATH, ".//div[contains(@class, 'w20 tl pr50')]").text.strip()
    except:
        match_info['Match'] = 'N/A'

    # Extract venue
    try:
        match_info['Inng'] = match.find_element(By.XPATH, ".//span[contains(@class, 'vn-venueDet')]").text.strip()
    except:
        match_info['Inng'] = 'N/A'

    try:
        match_info['NO'] = match.find_element(By.XPATH, ".//div[contains(@class, 'st-table__head')]").text.strip()
    except:
        match_info['NO'] = 'N/A'
    
    try:
        match_info['HS'] = match.find_element(By.XPATH, ".//div[contains(@class, 'st-table__head')]").text.strip()
    except:
        match_info['HS'] = 'N/A'

    try:
        match_info['Avg'] = match.find_element(By.XPATH, ".//div[contains(@class, 'st-table__head')]").text.strip()
    except:
        match_info['Avg'] = 'N/A'

    try:
        match_info['BF'] = match.find_element(By.XPATH, ".//div[contains(@class, 'st-table__head')]").text.strip()
    except:
        match_info['BF'] = 'N/A'

    try:
        match_info['SR'] = match.find_element(By.XPATH, ".//div[contains(@class, 'st-table__head')]").text.strip()
    except:
        match_info['SR'] = 'N/A'

    try:
        match_info['100'] = match.find_element(By.XPATH, ".//div[contains(@class, 'st-table__head')]").text.strip()
    except:
        match_info['100'] = 'N/A'
    
    try:
        match_info['50'] = match.find_element(By.XPATH, ".//div[contains(@class, 'st-table__head')]").text.strip()
    except:
        match_info['50'] = 'N/A'

    try:
        match_info['4s'] = match.find_element(By.XPATH, ".//div[contains(@class, 'st-table__head')]").text.strip()
    except:
        match_info['4s'] = 'N/A'

    try:
        match_info['6s'] = match.find_element(By.XPATH, ".//div[contains(@class, 'st-table__head')]").text.strip()
    except:
        match_info['6s'] = 'N/A'


    print(match_info)  # Debugging print
    match_details.append(match_info)

# Save data to CSV
with open('match_2022orangecap.csv', mode='w', newline='', encoding='utf-8') as file:
    writer = csv.DictWriter(file, fieldnames=["POS", "Player", "Runs", "Match", "Inng", "NO", "HS", "Avg", "BF", "SR", "100", "50", "4s", "6s"])
    writer.writeheader()
    for match in match_details:
        writer.writerow(match)

# Close browser
driver.quit()

print("✅ Data saved to match_2023_dshfk.csv")
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import csv

# Setup WebDriver
options = webdriver.ChromeOptions()
options.headless = False  # Set to True for headless mode
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

# Open IPL stats page
driver.get("https://www.iplt20.com/stats/2022")

# Wait for the table container to load
wait = WebDriverWait(driver, 30)

# Scroll to trigger table loading
driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
time.sleep(3)  # Allow time for content to load

# Wait until the table is visible
table = wait.until(EC.visibility_of_element_located((By.CSS_SELECTOR, "table.st-table.statsTable")))

# Get all rows
rows = driver.find_elements(By.CSS_SELECTOR, "table tbody tr")

match_details = []
for row in rows:
    columns = row.find_elements(By.TAG_NAME, "td")
    
    if len(columns) < 5:  # Ensure row has data
        continue

    match_info = {
        "POS": columns[0].text.strip(),
        "Player": columns[1].text.strip(),
        "Runs": columns[2].text.strip(),
        "Mat": columns[3].text.strip(),
        "Inng": columns[4].text.strip(),
    }

    print(match_info)  # Debugging print
    match_details.append(match_info)

# Save data to CSV
with open('ipl_stats_2022.csv', mode='w', newline='', encoding='utf-8') as file:
    writer = csv.DictWriter(file, fieldnames=match_details[0].keys())
    writer.writeheader()
    writer.writerows(match_details)

# Close browser
driver.quit()

print("✅ Data saved to ipl_stats_2022.csv")
```
